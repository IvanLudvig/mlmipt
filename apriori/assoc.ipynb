<h1>Apriori with apyori</h1>

```python
import pandas as pd
import apyori
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv')
```

```python
df.head()
```

```python
df = df.ffill(axis=1)
df.head()
```

```python
transactions = []
for i in range(0, len(df)):
    transactions.append([str(df.values[i, j]) for j in range(0, 20)])
```

```python
result = list(apyori.apriori(transactions, min_support = 0.004, min_confidence = 0.3, min_lift = 3, min_length = 2))
```

```python
result[0]
```

```python
columns = ['items', 'items_base', 'items_add', 'confidence', 'lift', 'support']
res = []
for row in result:
    for stats in row.ordered_statistics:
        a = []
        a.append(row.items)
        a.append(stats.items_base)
        a.append(stats.items_add)
        a.append(stats.confidence)
        a.append(stats.lift)
        a.append(row.support)
        res.append(a)
```

```python
df = pd.DataFrame(res, columns=columns)
```

```python
df.sort_values('support', ascending=False)
```
